### Generate data quality suggestions with AutoViz

```python
# pip install autoviz
# pip install jinja2==3.1.2
# pip install jupyter_bokeh
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
report = AV.AutoViz("houses.csv")
```

### You can also generate common charts with AutoViz

```python
# to see the visualizations instead, try this
# remember to change the target variable based on your own dataset
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()

# depVar => dependent variable => target variable
report = AV.AutoViz("houses.csv", chart_format="bokeh", depVar="price")
```
