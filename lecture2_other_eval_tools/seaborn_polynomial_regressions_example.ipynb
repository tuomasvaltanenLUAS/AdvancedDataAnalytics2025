### Using seaborn polynomial regression plots to study the complexity of the data

This can be useful while selecting your machine learning model, or especially if you're optimizing an algorithm, for example, with PolynomialFeatures.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("houses.csv")

# init plot
fig, axes = plt.subplots(1, 4, figsize=(20, 4))

# compared variables
var1 = "grade"
var2 = "price"

# title
fig.suptitle(f'Polynomial regression plot - ({var1}, {var2}, 2-5 degrees)')

# create plots
sns.regplot(ax=axes[0], x=var1, y=var2, data=df[[var1, var2]], order=2)
sns.regplot(ax=axes[1], x=var1, y=var2, data=df[[var1, var2]], order=3)
sns.regplot(ax=axes[2], x=var1, y=var2, data=df[[var1, var2]], order=4)
sns.regplot(ax=axes[3], x=var1, y=var2, data=df[[var1, var2]], order=5)
```
